```
import pandas as pd
df=pd.read_csv("/content/fb.csv.txt",parse_dates=['Date'],index_col='Date')
df
```

```
df.shift(2)#shifts the price by 2
```

```
df.shift(-1)
```

```
df['Prevoius day column']=df['Price'].shift(1)
df
```

```
df['1 day change']=df['Price']-df['Prevoius day column']
df
```

```
df.tshift(1) #shifts the index by 1
```
